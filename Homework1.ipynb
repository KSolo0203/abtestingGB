<h1>Задание 1. Выберите любой продукт.</h1>
Например, ваш продукт:<br>
- ozon.ru<br>
- citilink.ru<br>
- telegram<br>
Какие бы ключевые метрики вы выделили? Опишите, почему вы выбрали именно их и можно ли их измерить.<br>

Для маркетплейсов OZON и Citilink важными являются DAU, WAU, MAU, т.к. между количеством посетителей и выручкой существует прямая зависимость.<br>В Telegram существует премиальная подписка, следовательно существует и CR по данному признаку. Кроме того, для премиальных пользователей можно измерить CAC, ARPU, LTV (разница не ясна) и churn rate.<br>
Для баннеров на сайте считаются клики (по количеству формируемых запросов на сервер?).

<h1>Задание 2. Сформулируйте 5-10 гипотез для выбранного продукта.</h1>
OZON:<br>
- Следует провести рекламную кампанию по привлечению продавцов товара из сегмента A, т.к. это значительно повысит конкуренцию, что приведет к снижению цен в данном сегменте на платформе, что в свою очередь привлечет B новых клентов.<br>
- Следует увеличить максимальный срок хранения товара в ПВЗ на 1 сутки, что позволит увеличить CR в заказ на X%, т.к. процесс приобретения товара становится более гибким.<br>
Citilink:<br>
- Следует внедрить рекомендательную систему, подбирающую аналогичный товар от нескольких производителей, в т.ч. и более дорогостоящий, что позволит увеличить выручку на X рублей.<br>
- Следует провести смс и e-mail рассылку пользователям, когда-либо совершавшим покупки на платформе, информирующую о новых товарах в целевых сегментах, что позволит увеличить retention rate на X%.<br>
Telegram:<br>
- Следует внедрить изменение в дизайне мессенджера, позволяющие премиальным пользователям одним кликом отправлять промо-код на подписку своим контактам, что позволит увеличить CR на X%.
